## Data cleaning

### Subtask:
Clean the salary data and handle missing values in the dataset.


**Reasoning**:
Create the 'Salary' column with a placeholder value, handle missing values in other columns, and standardize 'Job Title' and 'Location' columns.



```python
# Handle missing salaries
df['Salary'] = 0  # Placeholder value for salary
df['Salary_Available'] = df['Salary From'].notnull() # Add a flag indicating the presence of salary information

# Address missing values in other columns
# For columns with a low percentage of missing values, impute with the median.
for col in ['Employees', 'Industry', 'Company Name', 'Employer Email', 'Employer Website', 'Employer Phone', 'Companydescription', 'Employer Location', 'Employer City', 'Employer State', 'Employer Zip Code']:
    if df[col].isnull().sum() < len(df) * 0.1:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0])
    else:
        # For columns with high percentage of missing values, fill with "Unknown"
        df[col] = df[col].fillna('Unknown')


# Address inconsistent job titles and locations
df['Job Title'] = df['Job Title'].str.lower().str.strip() #Lowercase and remove whitespace from job titles.
df['Location'] = df['Location'].str.lower().str.strip() #Lowercase and remove whitespace from locations.

# Data Validation
print("Number of missing salaries:", df['Salary'].isnull().sum())
print("Number of missing values after imputation:\n", df.isnull().sum())
print("\nUnique Job Titles (first 20):\n", df['Job Title'].unique()[:20])
print("\nUnique Locations (first 20):\n", df['Location'].unique()[:20])
```

## Data wrangling

### Subtask:
Create new features from existing ones in the dataframe.


**Reasoning**:
Extract keywords from the 'Job Description' column, create 'Minimum Salary' and 'Maximum Salary' columns, and convert the 'Salary' column to numeric type.



```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Extract keywords from 'Job Description'
def extract_keywords(text):
    if isinstance(text, str):
        # Basic preprocessing (remove HTML tags, lowercase, etc.)
        text = re.sub('<.*?>', '', text)  # Remove HTML tags
        text = text.lower()
        # Use TF-IDF to find important words (example)
        vectorizer = TfidfVectorizer(stop_words='english')
        tfidf_matrix = vectorizer.fit_transform([text])
        feature_names = vectorizer.get_feature_names_out()
        # Get top keywords (e.g., top 5)
        top_keywords = [feature_names[i] for i in tfidf_matrix.toarray().argsort()[0][-5:]]
        return ', '.join(top_keywords)
    else:
        return ''

df['Keywords'] = df['Job Description'].apply(extract_keywords)

# 2. Create 'Minimum Salary' and 'Maximum Salary' columns
def split_salary(salary_string):
    if pd.isna(salary_string) or not isinstance(salary_string, str):
        return 0, 0
    # Use regular expression to find salary values
    match = re.findall(r"(\d+)", salary_string)
    if len(match) == 2:
      return int(match[0]), int(match[1])
    elif len(match) == 1:
      return int(match[0]), int(match[0])
    else:
      return 0, 0

df['Minimum Salary'], df['Maximum Salary'] = zip(*df['Salary'].apply(split_salary))

# 3. Convert salary columns to numeric
# Salary has already been converted to numeric when the 0 value was filled in.

display(df.head())
```

## Feature engineering

### Subtask:
Develop features for matching resumes to jobs based on the provided data.  Since resume data isn't directly provided, focus on creating features from the job descriptions.


**Reasoning**:
Extract skills using regular expressions and TF-IDF, then vectorize job descriptions and extracted skills using TF-IDF.



```python
import re
from sklearn.feature_extraction.text import TfidfVectorizer

def extract_skills(text):
    if isinstance(text, str):
        # Basic preprocessing
        text = re.sub('<.*?>', '', text)
        text = text.lower()

        # Skill extraction using regular expressions
        skills_regex = r"\b(communication|leadership|problem-solving|teamwork|project management|python|java|sql|machine learning|data analysis)\b"
        extracted_skills = re.findall(skills_regex, text)

        # Combine extracted skills and TF-IDF keywords
        vectorizer = TfidfVectorizer(stop_words='english', max_features=100)
        tfidf_matrix = vectorizer.fit_transform([text])
        feature_names = vectorizer.get_feature_names_out()
        top_keywords = [feature_names[i] for i in tfidf_matrix.toarray().argsort()[0][-5:]]

        all_skills = list(set(extracted_skills + top_keywords))
        return ', '.join(all_skills)
    else:
        return ''

df['Extracted_Skills'] = df['Job Description'].apply(extract_skills)

# TF-IDF vectorization for 'Job Description'
vectorizer_desc = TfidfVectorizer(stop_words='english', max_features=500, ngram_range=(1, 2))
tfidf_matrix_desc = vectorizer_desc.fit_transform(df['Job Description'])

# TF-IDF vectorization for 'Extracted_Skills'
vectorizer_skills = TfidfVectorizer(stop_words='english', max_features=200)
tfidf_matrix_skills = vectorizer_skills.fit_transform(df['Extracted_Skills'])

print(tfidf_matrix_desc.shape)
print(tfidf_matrix_skills.shape)
```

**Reasoning**:
Combine the TF-IDF matrices into a single feature matrix by concatenating them horizontally.



```python
import scipy.sparse as sparse

# Combine TF-IDF matrices
combined_features = sparse.hstack([tfidf_matrix_desc, tfidf_matrix_skills])
print(combined_features.shape)
```

## Data splitting

### Subtask:
Split the combined features and the target variable into training and testing sets.


**Reasoning**:
Split the combined features and target variable into training and testing sets using train_test_split.



```python
from sklearn.model_selection import train_test_split

# Define the target variable (Job Title)
y = df['Job Title']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(combined_features, y, test_size=0.2, random_state=42)
```

## Model training

### Subtask:
Train a RandomForestClassifier model to predict job titles based on the engineered features.


**Reasoning**:
Train a RandomForestClassifier model to predict job titles based on the engineered features.



```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the model
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Train the model
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)
```

## Model evaluation

### Subtask:
Evaluate the performance of the trained RandomForestClassifier model.


**Reasoning**:
Evaluate the performance of the trained RandomForestClassifier model using various metrics.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

# Generate classification report
print(classification_report(y_test, y_pred, zero_division=0))

print("Analysis of Model Performance:")
print("The model's accuracy, precision, recall, and F1-score are presented above.  Due to the lack of salary information in the dataset, the model cannot incorporate salary into the job recommendations.")
print("The model's performance on specific job titles can be evaluated by examining the classification report.  Further investigation could focus on improving the model's performance on job titles that are currently not predicted accurately.")
```
